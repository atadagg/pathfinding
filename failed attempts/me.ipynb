```python
from tkinter import Tk, Canvas
import random
import queue

canvas_height = 450
canvas_width = 750


nums = queue.Queue()
nums.put("")
add = ""
previous_moves = []
cikmaz = []
cin = True


def get_maze(maze):
    with open(maze) as f:
        lines = f.read().splitlines()
        return [line.strip() for line in lines]

def get_robot_xy():
    for y, row in enumerate(maze):
        x = row.find("R")
        if(x != -1):
            return x, y

def draw_maze():
    for y, row in enumerate(maze):
        for x, clm in enumerate(row):
            if(is_wall(x, y)):
                draw_wall(x, y)


def draw_wall(x, y):
    draw(x, y, "black")

def draw_robot(x, y):
    draw(x, y, "orange")

def draw_footprint(x, y):
    draw(x, y, "grey")

def draw(x,y,color):

    maze_height = len(maze)
    maze_width = len(maze[1])
    height = canvas_height/maze_height
    width = canvas_width/maze_width

    x1, y1 = (x*width, y*height)
    x2, y2 = (x1+width, y1+height)
    w.create_rectangle(x1, y1, x2, y2, fill=color, outline=color)

def is_wall(x, y):
    return maze[y][x] == '#'

def is_exit(x, y):
    return maze[y][x] == 'E'

def move(direction):
    global x,y
    if direction == "R" and is_wall(x+1,y) == False:
        x += 1
        return x, y
    elif direction == "L" and is_wall(x-1,y) == False:
        x -= 1
        return x, y
    elif direction == "D" and is_wall(x,y-1) == False:
        y -= 1
        return x, y
    elif direction == "U" and is_wall(x,y+1) == False:
        y += 1
        return x, y
    else:
        return False

def go_back():
    global x,y,previous_moves
    try:
        x = previous_moves[-1][0]
        y = previous_moves[-1][1]
        previous_moves.pop()
        cikmaz.append([x,y])
    except:
        pass
def solve():
    global root,x,y,cin,cikmaz,i
    i = 0
    add = nums.get()
    for j in ["R","L","D","U"]:
        put = add + j
        for cx,cy in cikmaz:
            if cx == x and cy == y:
                cin = False
        if cin:
            if i == 0 and is_wall(x+1,y) == False:
                nums.put(put)
                draw_footprint(x,y)
                previous_moves.append([x,y])
                move(j)
                draw_robot(x,y)
                break
            elif i == 1 and is_wall(x-1,y) == False:
                nums.put(put)
                draw_footprint(x,y)
                previous_moves.append([x,y])
                move(j)
                draw_robot(x,y)
                break
            elif i == 2 and is_wall(x,y-1) == False:
                nums.put(put)
                draw_footprint(x,y)
                previous_moves.append([x,y])
                move(j)
                draw_robot(x,y)
                break
            elif i == 3 and is_wall(x+1,y+1) == False:
                nums.put(put)
                draw_footprint(x,y)
                previous_moves.append([x,y])
                move(j)
                draw_robot(x,y)
                break
            i += 1
        else:
            go_back()
            cin = True

    root.after(10, solve) ### really important

def main():
    global root,w,maze,x,y
    maze = get_maze("mymaze.txt")
    x, y = get_robot_xy()
    root = Tk()
    root.title("Map")
    w = Canvas(root, bg="white", width=canvas_width, height=canvas_height)
    w.pack()
    draw_maze()
    solve()
    root.mainloop()




main()
```
